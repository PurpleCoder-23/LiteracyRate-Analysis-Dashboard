# **DATA TRANSFORMATION**

####**LiteracyRate=Numberofliteratepersonsinagivenagegroup/Total Population in the same age group 100%**





```python
# renaming columns for ease of understanding and convenience

df=df.rename(columns={"Literacy Rate (Persons) - Total - 2001":"Total Literacy Rate in 2001",
                      "Literacy Rate (Persons) - Total - 2011":"Total Literacy Rate in 2011",
                      "Literacy Rate (Persons) - Rural - 2001":"Rural Literacy Rate in 2001",
                      "Literacy Rate (Persons) - Rural - 2011":"Rural Literacy Rate in 2011",
                      "Literacy Rate (Persons) - Urban - 2001":"Urban Literacy Rate in 2001",
                      "Literacy Rate (Persons) - Urban - 2011":"Urban Literacy Rate in 2011"})
```

```python
df.head(1)
```

```python
df.columns
```

###### Calculating growth rate to quantify the actual change over the decade 2001-2011,  This measures the percentage change over the ten-year period.                                                                                                                                                                        **GrowthRate**=( **LiteracyRate2011LiteracyRate2001**/**Literacy Rate in 2001)100**%



```python
#Creating new columns for growth rate

df['Total Growth Rate'] = ((df['Total Literacy Rate in 2011'] - df['Total Literacy Rate in 2001']) / df['Total Literacy Rate in 2001']) * 100
df['Rural Growth Rate'] = ((df['Rural Literacy Rate in 2011'] - df['Rural Literacy Rate in 2001']) / df['Rural Literacy Rate in 2001']) * 100
df['Urban Growth Rate'] = ((df['Urban Literacy Rate in 2011'] - df['Urban Literacy Rate in 2001']) / df['Urban Literacy Rate in 2001']) * 100
```

#**Data Visualization**

```python

# 1. Calculate the counts for each category
category_counts = df['Category'].value_counts() # .value_counts() is a concise way to get counts

# 2. Prepare data for the pie chart
# The labels will be the category names, and the sizes will be their counts.
labels = category_counts.index
sizes = category_counts.values

colors = sns.palettes.mpl_palette('Dark2', n_colors=len(labels))

# 3. Create the pie chart
plt.figure(figsize=(4, 4))
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct='%.1f%%', # This formats the percentage to one decimal place
    startangle=90,     # Starts the first slice at the top (12 o'clock)
    pctdistance=0.85,  # Distance of the percentage labels from the center
    wedgeprops={'edgecolor': 'white', 'linewidth': 1} # Adds white lines between slices
)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Set title
plt.title('Distribution of Categories', fontsize=16)

plt.tight_layout() # Adjust layout
plt.show()
```

```python
# Lets Visualize the distribution of Categories with a horizontal Bar Chart




# Calculate the counts for each category
category_counts = df.groupby('Category').size()

# Create the horizontal bar plot
ax = category_counts.plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))

# Remove the top and right spines
plt.gca().spines[['top', 'right']].set_visible(False)

# Add exact numbers as text labels
for index, value in enumerate(category_counts):
    # 'value' is the count, 'index' is the position of the bar on the y-axis
    ax.text(value, index, str(value), va='center', ha='left', fontsize=10, color='black')

# Set title and labels for clarity (optional, but recommended)
plt.title('Count of Each Category')
plt.xlabel('Count')
plt.ylabel('Category')

plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```

**TOTAL LITERACY RATE(2001-2011) ANALYSIS**

```python

# Sort the data by a relevant column (2011 literacy rate) for better visual flow
df_sorted = df.sort_values(by="Total Literacy Rate in 2011", ascending=True).reset_index(drop=True)

# Melt the dataframe to long format for easier plotting with seaborn
df_melted = df_sorted.melt(
    id_vars=["Country/ States/ Union Territories Name"],
    value_vars=["Total Literacy Rate in 2001", "Total Literacy Rate in 2011"],
    var_name="Year",
    value_name="Total Literacy Rate"
)
# Replace the column names in the 'Year' column for cleaner labels in the plot
df_melted['Year'] = df_melted['Year'].replace({
    "Total Literacy Rate in 2001": "2001",
    "Total Literacy Rate in 2011": "2011"
})

```

```python
df_melted.head()
```

```python
# Create the grouped horizontal bar chart

plt.figure(figsize=(14, 25))
ax = sns.barplot(
    x="Total Literacy Rate",  # X-axis will be the numerical literacy rate
    y="Country/ States/ Union Territories Name", # Y-axis will be the state names
    hue="Year", # Hue separates bars by year
    data=df_melted,
    palette='viridis'
)

# Set X-axis limits for percentages
plt.xlim(0, 100)

# Add exact numbers as text labels on the bars with percentage sign
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', padding=3, fontsize=8, color='black')

# Get current y-axis tick labels
current_labels = ax.get_yticklabels()

# Modify labels to add a newline character for visual spacing
new_labels = [label.get_text() + '\n \n \n \n' for label in current_labels]

# Set the modified labels back to the y-axis
ax.set_yticklabels(new_labels)

# Set plot title and labels
plt.title('Total Literacy Rate in Indian States/UTs (2001 vs 2011)', fontsize=16)
plt.xlabel('Total Literacy Rate (%)', fontsize=12)
plt.ylabel('State/Union Territory Name', fontsize=12)

# Remove the top and right spines for a cleaner look
plt.gca().spines[['top', 'right']].set_visible(False)

# Add a legend
plt.legend(title='Year', fontsize=10, title_fontsize=12, bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)

# Adjust layout to prevent labels from overlapping
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import json
import pandas as pd
import geopandas as gpd
import plotly.express as px

# Load the GeoJSON file for India states boundaries
url = "https://raw.githubusercontent.com/Subhash9325/GeoJson-Data-of-Indian-States/master/Indian_States"
gdf = gpd.read_file(url)

# Reduce complexity of geometry for better performance in plotting
gdf["geometry"] = gdf.to_crs(gdf.estimate_utm_crs()).simplify(1000).to_crs(gdf.crs)

# Rename the state column to 'ST_NM' to match common GeoJSON properties for states
india_states = gdf.rename(columns={"NAME_1": "ST_NM"}).__geo_interface__

data=df

# Rename the state column in  original df to match the geojson featureidkey
df_density = df.rename(columns={'Country/ States/ Union Territories Name': 'ST_NM'}).copy()

# Create the Choropleth map using Plotly Express
fig_density = px.choropleth(
    df_density,
    geojson=india_states,
    featureidkey='properties.ST_NM',  # Key in geojson that matches locations
    locations='ST_NM',                  # Column in df_density that matches featureidkey
    color='Total Literacy Rate in 2011', # Column to color the map by
    color_continuous_scale="Plasma",    # Choose a color scale
    range_color=(df_density['Total Literacy Rate in 2011'].min(), df_density['Total Literacy Rate in 2011'].max()), # Set color range
    scope="asia",                       # Set scope to Asia for a closer view
    labels={'Total Literacy Rate in 2011': 'Total Literacy Rate (2011)'} # Label for the color bar
)

# Customize the map layout
fig_density.update_geos(
    visible=True,
    projection_type="mercator",
    fitbounds="locations"
)
fig_density.update_layout(
    title_text='Density Map of Total Literacy Rate in India (2011)',
    title_x=0.5,                      # Center the title
    margin={"r": 0, "t": 40, "l": 0, "b": 0}  # Adjust margins
)


fig_density.show()
```
